### Importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing DataSet

```python
data = pd.read_csv(r"C:\Users\AI\Downloads\covid19_Confirmed_dataset.csv")
data.head()
```

### Checking Shape of Data

```python
data.shape
```

### Deleting the useless columns

```python
data.drop(['Long','Lat'],axis=1,inplace=True)
```

```python
data.head(15)
```

### Aggregating the rows by country

```python
df = data.groupby('Country/Region').sum()
```

```python
df
```

### Visualizing data related to a country
Visulalization always helps in better understanding of data

```python
df.loc['China'].plot()
df.loc['Italy'].plot()
df.loc['Spain'].plot()
df.loc['India'].plot()
plt.legend()
```

#### Finding first derivative and plotting the curve

```python
df.loc['China'].diff()
```

```python
df.loc['China'].diff().plot()
```

```python
df.loc['India'].diff().plot()
```

#### Finding maximum infection rate of each country

```python
df.loc['China'].diff().max()
```

```python
country = list(df.index)
country
max_cases = []
```

```python
for c in country:
    max_cases.append(df.loc[c].diff().max())
df['Max_infection_cases']=max_cases
```

```python
df.head()
```

```python
df1 = pd.DataFrame(df['Max_infection_cases'])
```

#### Created a new data frame using only needed columns

```python
df1.head()
```
